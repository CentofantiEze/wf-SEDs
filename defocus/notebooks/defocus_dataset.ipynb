```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paths
train_dataset_path = '../../../output/defocus_dataset/train_Euclid_res_2000_TrainStars_id_010GT_350_bins_defocus.npy'
test_dataset_path = '../../../output/defocus_dataset/test_Euclid_res_id_010GT_350_bins_defocus.npy'
```

```python
# Load datasets
train_dataset = np.load(train_dataset_path, allow_pickle=True)[()]
test_dataset = np.load(test_dataset_path, allow_pickle=True)[()]
```

```python
print(train_dataset.keys(),'\n')
print(test_dataset.keys())
```

# Defocused stars

```python
star_id = 300
font_size=17
vmax=0.15
vmin=-0.003

plt.rcParams.update({
    "figure.facecolor":  (1.0, 1.0, 1.0, 1.0),  
    "axes.facecolor":    (1.0, 1.0, 1.0, 1.0),  
    "savefig.facecolor": (1.0, 1.0, 1.0, 1.0),  
})

fig, ax = plt.subplots(1,3, figsize=(15,4), )

im = ax[0].imshow(train_dataset['negative_defocused_stars'][star_id], cmap='gist_stern',vmax=vmax, vmin=vmin)
ax[0].set_xticks([])
ax[0].set_yticks([])
plt.colorbar(im, ax=ax[0])
ax[0].set_title('Negative defocus', fontsize=font_size)

im = ax[1].imshow(train_dataset['noisy_stars'][star_id], cmap='gist_stern',vmax=vmax, vmin=vmin)
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.colorbar(im, ax=ax[1])
ax[1].set_title('On focus', fontsize=font_size)

im = ax[2].imshow(train_dataset['positive_defocused_stars'][star_id], cmap='gist_stern',vmax=vmax, vmin=vmin)
ax[2].set_xticks([])
ax[2].set_yticks([])
plt.colorbar(im, ax=ax[2])
ax[2].set_title('Positive defocus', fontsize=font_size)

plt.show()
```
