```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# optimisation history path
optim_hist_path = '../wf-outputs/optim-hist/optim_hist_poly_interp_%s_sigma_%s.npy'
```

```python
# Load optimisation history
n_dataset_list = [('8bins','0'), ('8bins','005'),('8bins','01'),('8bins','02')]
optim_hist_list = [np.load(optim_hist_path % n_dataset, allow_pickle=True)[()] for n_dataset in n_dataset_list]
```

```python
plot_loss = 'val_loss'

print('8 bin SEDs model - optimisation loss')

plt.figure(figsize=(14,4))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['param_cycle1'][plot_loss], label='sigma = {}'.format(n_dataset_list[idx][1]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,4))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['nonparam_cycle1'][plot_loss], label='sigma = {}'.format(n_dataset_list[idx][1]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,4))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['param_cycle2'][plot_loss], label='sigma = {}'.format(n_dataset_list[idx][1]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,4))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['nonparam_cycle2'][plot_loss], label='sigma = {}'.format(n_dataset_list[idx][1]))
plt.yscale('log')
plt.legend()
plt.show()
```
