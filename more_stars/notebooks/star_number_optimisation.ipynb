```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# optimisation history path
optim_hist_path = '../wf-outputs/optim-hist/optim_hist_poly_full_poly_{}_stars.npy'
```

```python
# Load optimisation history
optim_hist_1000 = np.load(optim_hist_path.format(1000), allow_pickle=True)[()]
optim_hist_2000 = np.load(optim_hist_path.format(2000), allow_pickle=True)[()]
optim_hist_6000 = np.load(optim_hist_path.format(6000), allow_pickle=True)[()]
optim_hist_8000 = np.load(optim_hist_path.format(8000), allow_pickle=True)[()]
optim_hist_10000 = np.load(optim_hist_path.format(10000), allow_pickle=True)[()]

n_stars_list = [500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
optim_hist_list = [np.load(optim_hist_path.format(n_stars), allow_pickle=True)[()] for n_stars in n_stars_list]
```

```python
print(optim_hist_6000.keys())
print(optim_hist_6000['param_cycle1'].keys())
```

```python
plot_loss = 'val_loss'
optim_hist_6000_loss = optim_hist_6000['param_cycle1'][plot_loss]+optim_hist_6000['nonparam_cycle1'][plot_loss]+(optim_hist_6000['param_cycle2'][plot_loss])+(optim_hist_6000['nonparam_cycle2'][plot_loss])
optim_hist_8000_loss = optim_hist_8000['param_cycle1'][plot_loss]+optim_hist_8000['nonparam_cycle1'][plot_loss]+(optim_hist_8000['param_cycle2'][plot_loss])+(optim_hist_8000['nonparam_cycle2'][plot_loss])
optim_hist_10000_loss = optim_hist_10000['param_cycle1'][plot_loss]+optim_hist_10000['nonparam_cycle1'][plot_loss]+(optim_hist_10000['param_cycle2'][plot_loss])+(optim_hist_10000['nonparam_cycle2'][plot_loss])
```

```python
plt.figure(figsize=(14,8))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['param_cycle1'][plot_loss], label='{} stars'.format(n_stars_list[idx]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,8))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['nonparam_cycle1'][plot_loss], label='{} stars'.format(n_stars_list[idx]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,8))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['param_cycle2'][plot_loss], label='{} stars'.format(n_stars_list[idx]))
plt.yscale('log')
plt.legend()
plt.show()

plt.figure(figsize=(14,8))
for idx,hist_stars in enumerate(optim_hist_list):
    plt.plot(hist_stars['nonparam_cycle2'][plot_loss], label='{} stars'.format(n_stars_list[idx]))
plt.yscale('log')
plt.legend()
plt.show()
```
